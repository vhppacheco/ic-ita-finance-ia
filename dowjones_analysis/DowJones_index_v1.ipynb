#Getting Dow Jones assets list

##Getting Dow Jones assets list from Business Insider

```python
url = 'https://markets.businessinsider.com/index/dow_jones'
#sleep(5)
html = requests.get(url).content
df_list = pd.read_html(html)
dowjones_bi = df_list[0]
```

```python
url = 'https://markets.ft.com/data/indices/tearsheet/constituents?s=DJI:DJI'
#sleep(5)
html = requests.get(url).content
df_list = pd.read_html(html)
dowjones_ft = df_list[0]
```

##Getting Dow Jones assets list from SlickCharts.com

```python
# Grab content from URL (Pegar conteúdo HTML a partir da URL)
url = "https://www.slickcharts.com/dowjones"

driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get(url)
driver.implicitly_wait(10)  # in seconds

element = driver.find_element_by_xpath(
        "/html/body/div[2]/div[3]/div[1]/div/div/table")
        
html_content = element.get_attribute('outerHTML')

# Parse HTML (Parsear o conteúdo HTML) - BeaultifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

# Data Structure Conversion (Estruturar conteúdo em um Data Frame) - Pandas
dowjones_slickcharts = pd.read_html(str(table))[0]
```

##Getting Dow Jones assets list from Stock Market MBA

```python
# Grab content from URL (Pegar conteúdo HTML a partir da URL)
url = "https://stockmarketmba.com/stocksinthedjia.php"

driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get(url)
driver.implicitly_wait(10)  # in seconds

element = driver.find_element_by_xpath(
        '''//*[@id="ipos"]''')
        
html_content = element.get_attribute('outerHTML')

# Parse HTML (Parsear o conteúdo HTML) - BeaultifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find(name='table')

# Data Structure Conversion (Estruturar conteúdo em um Data Frame) - Pandas
dowjones_stockmarketmba = pd.read_html(str(table))[0]
dowjones_stockmarketmba = dowjones_stockmarketmba[:len(dowjones_stockmarketmba)-1]
```
